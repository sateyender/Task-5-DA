```python
import seaborn as sns
import matplotlib.pyplot as plt

# =======================
# 📌 Univariate Analysis
# =======================

# Histogram for numeric features
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Histogram of Numeric Features", fontsize=16)
plt.show()

# Survival count
sns.countplot(x='Survived', data=df, palette='viridis')
plt.title("Survival Count")
plt.show()

# Gender distribution
sns.countplot(x='Sex', data=df, palette='mako')
plt.title("Gender Distribution")
plt.show()

# =======================
# 📌 Bivariate Analysis
# =======================

# Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df, palette='coolwarm')
plt.title("Survival Count by Gender")
plt.show()

# Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df, palette='viridis')
plt.title("Survival Count by Passenger Class")
plt.show()

# =======================
# 📌 Outlier Detection
# =======================

sns.boxplot(x=df['Fare'])
plt.title("Boxplot - Fare")
plt.show()

sns.boxplot(x=df['Age'])
plt.title("Boxplot - Age")
plt.show()

# =======================
# 📌 Correlation Analysis
# =======================

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# =======================
# 📌 Pairplot
# =======================

sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='husl')
plt.show()
```

```python
# 🛳️ Titanic Dataset – Exploratory Data Analysis (Final Colab Version)

# =======================
# 📌 Step 1: Import Libraries & Load Data
# =======================
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Display all columns
pd.set_option('display.max_columns', None)

# Load dataset directly from GitHub
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("✅ Titanic dataset loaded successfully!")

# =======================
# 📌 Step 1A: Data Understanding
# =======================
print("\nShape of dataset:", df.shape)
print("\nFirst 5 rows:\n", df.head())
print("\nInfo:\n")
print(df.info())
print("\nSummary statistics:\n", df.describe())

print("\nMissing values in each column:\n", df.isnull().sum())

# Categorical column value counts
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nValue counts for {col}:\n", df[col].value_counts())

# =======================
# 📌 Step 2: Visualizations
# =======================

# Histograms for numeric features
numeric_df = df.select_dtypes(include=['number'])
numeric_df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Histogram of Numeric Features", fontsize=16)
plt.show()

# Survival count
sns.countplot(x='Survived', data=df, palette='viridis')
plt.title("Survival Count")
plt.show()

# Gender distribution
sns.countplot(x='Sex', data=df, palette='mako')
plt.title("Gender Distribution")
plt.show()

# Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df, palette='coolwarm')
plt.title("Survival Count by Gender")
plt.show()

# Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df, palette='viridis')
plt.title("Survival Count by Passenger Class")
plt.show()

# Outlier detection - Fare
sns.boxplot(x=df['Fare'])
plt.title("Boxplot - Fare")
plt.show()

# Outlier detection - Age
sns.boxplot(x=df['Age'])
plt.title("Boxplot - Age")
plt.show()

# Correlation heatmap (numeric only)
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Pairplot (numeric subset)
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='husl')
plt.show()

# =======================
# 📌 Step 3: Observations & Summary
# =======================
print("""
📌 OBSERVATIONS FROM TITANIC EDA:

1️⃣ Gender & Survival:
   - Females have a much higher survival rate than males.
   - Males make up the majority of passengers but had lower survival chances.

2️⃣ Passenger Class:
   - 1st Class passengers had the highest survival rate.
   - 3rd Class passengers had the lowest survival rate.

3️⃣ Age Factor:
   - Survival rate is higher among younger passengers.
   - Most passengers are between 20–40 years old.

4️⃣ Fare Insights:
   - Fare distribution is right-skewed (few passengers paid very high fares).
   - Higher fares are correlated with higher survival.

5️⃣ Missing Data:
   - 'Cabin' column has many missing values.
   - 'Age' also has some missing values but far fewer than 'Cabin'.

6️⃣ Correlation Highlights:
   - Fare has a mild positive correlation with survival.
   - Pclass is strongly linked to survival (lower class number = higher chance).

📤 NEXT STEPS:
   - Handle missing values before modeling.
   - Encode categorical variables for ML.
   - Engineer features like Family size, Title from Name.
""")
```
